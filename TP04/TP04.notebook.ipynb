```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
```

# Exercise 1

```python
X = np.array([[4, 1], [1, 2], [9, 14], [7, 10], [5, 16], [12, 14], [18, 10], [8, 10]])
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter plot of data')
plt.show()
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
def knn(X, point, k=4):
    distances = []
    for x in X:
        distances.append(euclidean_distance(point, x))
    indices = np.argsort(distances)[:k]
    return indices
```

```python
test_point = np.array([11, 3])
```

```python
k_nearest_neighbors_indices = knn(X, test_point)
```

```python
print("Indices of k nearest neighbors:", k_nearest_neighbors_indices)
```

```python
neigh = NearestNeighbors(n_neighbors=4)
neigh.fit(X)
distances, indices = neigh.kneighbors([test_point])
print("Indices of k nearest neighbors using sklearn:", indices)
```

# Exercise 2

```python
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_iris, y_iris)
```

```python
observations = np.array([[5.4, 3.3, 5.9, 1.1], [1, 1, 1, 1]])
```

```python
predicted_classes = knn_model.predict(observations)
print("Predicted classes:", predicted_classes)
```
